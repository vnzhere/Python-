Types of objects in python
1. immutable (value cannot be changed)
a. int
b. float
c. bool
d. tuple
e. strings

2. mutable (value can be changed)
a. list
b. set
c. dictionary

```python
#list []

my_list = [1,2,3,4,5]
print(my_list)
type(my_list)
```

```python
# can have multiple datatypes
my_list=[2, 3, 8 , 'english' , True]
print(my_list)
```

```python
#lists are mutable-------->> changeable

#add elements to a list

my_list=[2, 3, 8 , 'english' , True]
my_list.append(5)
print(my_list)
```

```python
# print elements of a list using their index

print(my_list[1])
print(my_list[4])
```

```python
# lists allow duplicate value

list_1= [1,2,3,4,5,6,7,8,2,3]
print(list_1)
print(len(list_1))
```

```python
#initiating an empty liest

list_2 =[]
print (list_2)

list_2.append(5)
print(list_2)
```
